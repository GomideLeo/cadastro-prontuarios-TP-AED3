```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../results.csv')
df = df.drop(df.columns[0], axis=1)
df['Tamanho do arquivo de dados'] = df['Tamanho do arquivo de dados'].astype('category')
df['Tamanho do arquivo Label'] = df['Tamanho do arquivo de dados'].map({10_000_024:'~10MB', 100_000_024:'~100MB', 250_000_024:'~250MB', 2_500_000_024:'~2GB'})
```

```python
df.info()
```

```python
# plt.rc(usetex=True)
f, axes = plt.subplots(1, 3, sharey=True)
# plt.Figure(figsize=(10,8))
f.set_figheight(5)
f.set_figwidth(15)
sns.barplot(y=df['Tempo de atualização de Registro'], x=df['Tamanho do arquivo Label'], hue=df['Tipo de Memória'], ax=axes[0])
sns.barplot(y=df['Tempo de atualização de Registro'], x=df['Profundidade Inicial do Diretório'], hue=df['Tipo de Memória'], ax=axes[1])
sns.barplot(y=df['Tempo de atualização de Registro'], x=df['Tamanho dos Buckets'], hue=df['Tipo de Memória'], ax=axes[2])

plt.savefig('update.svg')
```
